```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
apps = pd.read_csv("AppleStore.csv")
apps.head()
```

The fields are as follow:
* "id" : App ID
* "track_name": App Name
* "size_bytes": Size (in Bytes)
* "currency": Currency Type
* "price": Price amount
* "ratingcounttot": User Rating counts (for all version)
* "ratingcountver": User Rating counts (for current version)
* "user_rating" : Average User Rating value (for all version)
* "userratingver": Average User Rating value (for current version)
* "ver" : Latest version code
* "cont_rating": Content Rating
* "prime_genre": Primary Genre
* "sup_devices.num": Number of supporting devices
* "ipadSc_urls.num": Number of screenshots showed for display
* "lang.num": Number of supported languages
* "vpp_lic": Vpp Device Based Licensing Enabled

```python
apps.info()
```

```python
apps.describe(include='all')
```

* we remove unnecessary columns: unnamed, currency

```python
apps.drop("Unnamed: 0",axis=1,inplace=True)
apps.drop("currency",axis=1,inplace=True)
apps.head()
```

```python
# Games are the most common apps in the apple store
apps.prime_genre.value_counts()
```

```python
# user rating depending on the genre
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='prime_genre', y='user_rating')
ax.tick_params(axis='x', rotation=90)
print("Average user rating",apps.user_rating.mean())
print("max user rating",apps.user_rating.max())
print("min user rating",apps.user_rating.min())
```

```python
# user rating for free apps
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps[apps.price > 0], x='prime_genre', y='user_rating')
ax.tick_params(axis='x', rotation=90)
print("number of non-free apps",len(apps[apps.price > 0]))
print("Average user rating",apps[apps.price > 0].user_rating.mean())
print("max user rating",apps[apps.price > 0].user_rating.max())
print("min user rating",apps[apps.price > 0].user_rating.min())
```

```python
# user rating for free apps
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps[apps.price == 0], x='prime_genre', y='user_rating')
ax.tick_params(axis='x', rotation=90)
print("number of free apps",len(apps[apps.price == 0]))
print("Average user rating",apps[apps.price == 0].user_rating.mean())
print("max user rating",apps[apps.price == 0].user_rating.max())
print("min user rating",apps[apps.price == 0].user_rating.min())
```

```python
apps.cont_rating.value_counts()
```

```python
# user rating depending on the content rating
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='cont_rating', y='user_rating')
ax.tick_params(axis='x', rotation=90)
```

```python
# user rating depending on the number of languages
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='lang.num', y='user_rating')
ax.tick_params(axis='x', rotation=90)
# the number of languages has little to no effect on the user rating
```

```python
# user rating depending on the number of supported devices
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='sup_devices.num', y='user_rating')
ax.tick_params(axis='x', rotation=90)
# the number of supported devices has little to no effect on the user rating, mostly because most users use their phones
```

```python
# user rating depending on the number of languages
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='prime_genre', y='rating_count_tot')
ax.tick_params(axis='x', rotation=90)
# the number of supported devices has little to no effect on the user rating, mostly because most users use their phones
```

```python
# app prices in respect to genres
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(data=apps, x='prime_genre', y='price')
ax.tick_params(axis='x', rotation=90)
# it is noteworthy that the expansive apps are those that offer services that are not for entertainment (except for music)
# shopping apps are basically free
# finance, news, and social media apps are the least expansive
```

```python
ax = sns.lineplot(data=apps, x='price', y='rating_count_tot')
# the least expansive the app, the high rating count it has. this is natural as free apps have a large number of users, because it's free
```

```python
ax = sns.lineplot(data=apps, x='price', y='user_rating')
# it seems that more exansive apps always have high ratings, let's see how many apps are priced at over 50$
```

```python
# only 7 apps have prices of 50$ and over, compared to close to 8000 apps with less than 50$.
# hence, the above chart is misleading
apps[apps.price>=50].info()
```

```python
ax = sns.lineplot(data=apps[apps.price <50], x='price', y='user_rating')
# for apps with prices lower than 50$, the rating isn't impacted by the price, in fact, having a high price doesn't imply having a high rating
```
